```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
```

```python
# Trazer a média das vendas
media_vendas = round(df_faturamento['valor'].mean(), 2)
media_vendas
```

```python
# Ponto 2
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
df_faturamento['mes'] = df_faturamento.data_ref.dt.month_name()
df_faturamento.plot.bar(x='mes', y='valor', color='blue', figsize=(10, 5))
```

```python
# Criar um gráfico de linhas mostrando o mês de referência e o valor de faturamento
df_faturamento.plot(x='mes', y='valor', kind='line', color='red', figsize=(10, 5))
plt.title('Faturamento por Mês')
```
